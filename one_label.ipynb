```python
import json

# labels.txt to labels.json
with open('./labels.txt', 'r') as f:
    lines = f.readlines()
    
labelsdict = {}

for i, line in enumerate(lines):
    # Agate (玛瑙)\n -> Agate
    lines[i] = line.split()[0]
    labelsdict[str(i)] = lines[i]

mineral_dict = labelsdict
mineral_dict
```

```python
with open('./annotations/test.json', encoding="gbk") as f:
    samples = json.load(f)

one_samples = []

for sample in samples:
    if len(sample['labels']) == 1:
        one_samples.append(sample)

one_samples
```

```python
import random
import os

random.shuffle(one_samples)

n = len(one_samples)
print(n)

def sampls_2_json(samples, path):
    random.shuffle(samples)
    samples = json.dumps(samples, indent=4)
    with open(path, 'w') as f:
        f.write(samples)

dir = './annotations'
if not os.path.exists(dir):
    os.makedirs(dir)

sampls_2_json(one_samples, './annotations/one.json')
```
